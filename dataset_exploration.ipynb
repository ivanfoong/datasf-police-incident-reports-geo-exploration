```python
import pandas as pd

df = pd.read_csv("data/Police_Department_Incident_Reports__2018_to_Present_20240202.csv")
```

```python
df['Incident Datetime'] = pd.to_datetime(df['Incident Datetime'], format='%Y/%m/%d %H:%M:%S %p')
df['Report Datetime'] = pd.to_datetime(df['Report Datetime'], format='%Y/%m/%d %H:%M:%S %p')
```

```python
df_2022 = df[(df['Incident Datetime']>'2022-01-01') & (df['Incident Datetime']<'2023-01-01')]
df_2022['Incident Category'].unique()
```

```python
pd.options.display.max_rows = 4000
print(f"Total incidents reported to SF police in 2022: {df_2022.shape[0]}")
print()
print("Break down of incident categories and subcategories:")
df_2022.groupby(["Incident Category", "Incident Subcategory"])["Incident Description"].count()
```

```python
df_2022['Incident Category'] = df_2022['Incident Category'].str.replace('Motor Vehicle Theft?', 'Motor Vehicle Theft')
print(f"Total incidents reported to SF police in 2022: {df_2022.shape[0]}")
print()
print("Break down of incident categories and subcategories:")
df_2022.groupby(["Incident Category", "Incident Subcategory"])["Incident Description"].count()
```

```python
# plot using matplotlib
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
import contextily as ctx
import matplotlib.pyplot as plt

df_2022_motor_vehicle_theft = df_2022[df_2022['Incident Category'] == 'Motor Vehicle Theft']

print(f"There are {df_2022_motor_vehicle_theft.shape[0]} cases of reported Motor Vehicle Theft in SF 2022")

geometry = [Point(xy) for xy in zip(df_2022_motor_vehicle_theft['Longitude'], df_2022_motor_vehicle_theft['Latitude'])]
geo_df = GeoDataFrame(df_2022_motor_vehicle_theft, geometry=geometry)
geo_df.crs = "EPSG:4326"
geo_df = geo_df.to_crs(epsg=3857)

# Initialize our plot
fig, ax = plt.subplots(figsize=(10,10))
# Plot our SF GeoDataFrame
geo_df.plot(ax=ax, alpha = .1)
# Add in a background using contextily
ctx.add_basemap(ax)
# Turn off axis
ax.set_axis_off()
fig.tight_layout()
```
